```python
import os
import glob
from copy import deepcopy
import numpy as np
import pandas as pd
from PIL import Image
```

```python
DATASET_PATH = "D:\dataset\kaggle_FlowersRecognition"
TEST_RATE = 0.1
```

```python
path_class = os.listdir(DATASET_PATH)
class_data = []
```

```python
pd_list = []
for c in path_class:
    files = glob.glob(DATASET_PATH + '/' + c + '/*.jpg')
    pd_list.append(pd.DataFrame({'class' :[c for i in range(len(files))], 'path' : files}))
df = pd.concat(pd_list)
df.reset_index()
```

```python
#分割
df_s = df.sample(frac=1)
#分割点計算
dv_point = int(len(df_s) * TEST_RATE)
df_test = df_s[0:dv_point]
df_train = df_s[dv_point:]

#保存
df_test.to_csv("test_data.csv")
df_train.to_csv("train_data.csv")
```
